### Soal 02

Berapa banyak perenang yang memiliki medali perunggu(bronze)?

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('summer.csv')

df
```

```python
#your code goes here
# Data cek
df.info()
```

```python
# define filter untuk swimming dan medali bronze
rb = df.loc[(df["Discipline"] == "Swimming") & (df["Medal"] == "Bronze")] # rb = renang bronze
rb
```

```python
# hitung nilai masing-masing pada atlet dengan nilai unique
renang_bronze = rb["Athlete"].nunique()
```

Answer:


```python
# Answer
print("Banyak perenang yang memiliki medali perunggu (Bronze) adalah", renang_bronze, "Orang. ")
```

Dalam olahraga air (Aquatics), kota mana yang memiliki perolehan medali tertinggi?

```python
am = df.loc[df["Sport"] == "Aquatics"] # ab = aquatics medal
am
```

```python
# Gabung dengan kota dan dijumlahkan medalinya
aquatic_medals = am.groupby("City").size().reset_index(name="Total Medali")
aquatic_medals
```

```python
# define kota dengan medali tertinggi
kota_medali = aquatic_medals.loc[aquatic_medals["Total Medali"].idxmax()]["City"]
```

Answer:

```python
# Answer
print ("Dalam olahraga air (Aquatics), kota yang memiliki perolehan medali tertinggi adalah",(kota_medali),".")
```

### **Soal 3 List Comprehension & Pandas** 

Pada Soal ini kalian **wajib** menggunakan feature list comprehension dan juga apply.<br>

```python
import pandas as pd
df=pd.read_csv('Salaries.csv')
df.tail()
```

```python
df.info()
```

PT.A baru saja kena tegur dari negara karena ketauan tidak membayar uang lemburan. Anda sebagai seorang data Scientist diminta PT.A untuk mengolah data yang ada.

**Step pertama**

**Wajib Menggunakan .Apply()** <br>
PT.A melihat beberapa data tidak di isi oleh sebagian karyawan. kalian di minta untuk untuk merubah nilai 'Not provided' pada column `JobTitle` dan `EmployeeName` menjadi 'Error'.

```python
# menggunakan fungsi yang akan mencari nilai "not provided"
def fungsi_pengganti(nilai):
    if nilai == "Not provided":
        return "Error"
    return nilai
```

```python
# gunakan fungsi yang sudah di define tadi untuk kolom EmployeeName
df["EmployeeName"] = df["EmployeeName"].apply(fungsi_pengganti)
```

```python
# gunakan fungsi yang sudah di define tadi untuk kolom  JobTitle
df["JobTitle"] = df["JobTitle"].apply(fungsi_pengganti)
```

```python
df.tail() # cek perubahan
```

**Step Kedua**

**WAJIB Gunakan list comprehension**<br>
Anda diminta untuk menambahkan 200000 pada karyawan yang mendapat Overtime 0 pada Data yang telah dibersihkan di tahap 1.


```python
#your code goes here
# Menggunakan list comprehension untuk menambahkan 200000 untuk overtime karyawan yang bernilai 0
df["OvertimePay"] = [pay + 200000 if pay == 0 else pay for pay in df["OvertimePay"]]
```

```python
# answer
df
```

```python
#jangan tulis code disini
#cell contoh hasil:
```

**Tahap ketiga** 

**bebas memakai teknik apapun**<br>
selanjutnya ada kelasahan pada penulisan 'CAPTAIN III (POLICE DEPARTMENT)' ubahlah menjadi 'CAPTAIN III' saja

```python
# membuat fungsi yang akan mereplace POLICE DEPARTMENT dengan kosong " "
def hapus_pd(pd):
    return pd.replace(" (POLICE DEPARTMENT)", " ")
```

```python
# menggunakan fungsi tadi di dalam kolom JobTitle
df["JobTitle"] = df["JobTitle"].apply(hapus_pd)
```
