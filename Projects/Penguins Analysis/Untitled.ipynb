```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins').dropna()
penguins.head()
```

```python
penguins.info()
```

```python
# Uniqueness Categorical Variables
# Let's have a look at categorical variables. 
# How many unique values of these variables.

categorical = penguins.select_dtypes(["category", "object"]).columns
for cat_col in categorical:
    print(f"{cat_col} : {penguins[cat_col].nunique()} uniqueness variable(s)")
```

```python
# Discrete and Continuous Variables
# Let's have a look at Discrete and Continuous variables.

numeric = penguins.select_dtypes(["int", "float"]).columns
for num_col in numeric:
    print(f"{num_col} : {penguins[num_col].nunique()} uniqueness variable(s)")
```

```python
# Figure and Axes
fig, (ax1, ax2) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(18, 12))
# Scatterplot
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", ax=ax1)
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", ax=ax2)

# Title
ax1.set_title("Penguin Size", weight="bold", fontsize=30, pad=30)

# Labels
ax1.set_xlabel("bill length mm", weight="bold", fontsize=15, labelpad=10)
ax1.set_ylabel("bill depth mm", weight="bold", fontsize=15, labelpad=25)
ax2.set_xlabel("flipper length mm", weight="bold", fontsize=15, labelpad=10)
ax2.set_ylabel("body mass g", weight="bold", fontsize=15, labelpad=25)

# Ticklabels
ax1.set_xticklabels(labels=list(np.arange(30, 65, 5)),
                    weight="semibold",
                    fontsize=10)
ax1.set_yticklabels(labels=list(np.arange(12, 22, 2)),
                    weight="semibold",
                    fontsize=10)
ax2.set_xticklabels(labels=list(np.arange(160, 240, 10)),
                    weight="semibold",
                    fontsize=10)
ax2.set_yticklabels(labels=list(np.arange(2500, 6500, 500)),
                    weight="semibold",
                    fontsize=10);
```
