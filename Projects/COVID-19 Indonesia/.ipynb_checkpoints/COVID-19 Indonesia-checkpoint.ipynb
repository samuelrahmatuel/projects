```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df['Month-Year'] = df['Date'].dt.to_period('M')
df.head()
```

```python
grouped = df.groupby('Month-Year')[['Total Deaths',
                                    'Total Recovered',
                                    'Total Active Cases']].median().reset_index()
grouped
```

```python
grouped['Month-Year'] = grouped['Month-Year'].astype('str')
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'])
plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'])
plt.title("Deaths, Recovered & Active Cases of Covid 19")
plt.xlabel('Year-Month')
plt.ylabel('Person')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'])
plt.plot(grouped['Month-Year'], grouped['Total Recovered'])
plt.plot(grouped['Month-Year'], grouped['Total Active Cases'])

plt.title("Deaths, Recovered & Active Cases of Covid 19")
plt.xlabel('Year-Month')
plt.ylabel('Person')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'], 
         label ='Total Deaths')
plt.plot(grouped['Month-Year'], grouped['Total Recovered'], 
         label ='Total Recovered')
plt.plot(grouped['Month-Year'], grouped['Total Active Cases'], 
         label ='Total Active Cases')

plt.title("Deaths, Recovered & Active Cases of Covid 19")
plt.xlabel('Year-Month')
plt.ylabel('Person')
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'], 'k',
         label ='Total Deaths')
plt.plot(grouped['Month-Year'], grouped['Total Recovered'], 'r',
         label ='Total Recovered')
plt.plot(grouped['Month-Year'], grouped['Total Active Cases'], 'm',
         label ='Total Active Cases')

plt.title("Deaths, Recovered & Active Cases of Covid 19")
plt.xlabel('Year-Month')
plt.ylabel('Person')
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'], linestyle='--', color='#344e41',
         label ='Total Deaths')

plt.plot(grouped['Month-Year'], grouped['Total Recovered'], color='#590d22', linewidth = 3,
         label ='Total Recovered')

plt.plot(grouped['Month-Year'], grouped['Total Active Cases'], 'm',
         label ='Total Active Cases')

plt.title("Deaths, Recovered & Active Cases of Covid 19")
plt.xlabel('Year-Month')
plt.ylabel('Person')
plt.xticks(rotation=90)

plt.legend()
plt.show()
```
