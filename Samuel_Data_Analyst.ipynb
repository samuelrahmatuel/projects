# Soal 2
---
- Read data `sample_data_shopee.csv`
- Simpan dalam bentuk `Pandas DataFrame`
- Simpan dalam variabel `data`

```python
# Tulis code disini
data = pd.read_csv('https://raw.githubusercontent.com/samuelrahmatuel/projects/main/Datasets/data_shopee_sample.csv')
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
print(data.shape)
data.head()
```

# Soal 3
---
- Cari total `box_qty` dari seluruh data transaksi tersebut.
- Simpan dalam variabel `total_box_qty`

```python
# Tulis code disini
total_box_qty = data['box_qty'].sum()
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
print(f'Total box_qty: {total_box_qty}')
```

# Soal 4
---
- Cari rata-rata `box_qty` dari transaksi yang `total_weight`-nya lebih dari 10.
- Simpan dalam variabel `avg_box_qty`

```python
# Tulis code disini
avg_box_qty = data[data['total_weight'] > 10]['box_qty'].mean()
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
print(f'Rata-rata box_qty saat total-weight > 10: {avg_box_qty:.2f}')
```

# Soal 5
---
- Buat pivot table untuk menunjukkan total `box_qty` per customer name per area
- Simpan dalam variable `groupby_data`

```python
# Tulis code disini
groupby_data = data.pivot_table(values='box_qty',
                              index=['area', 'customer_name'],
                              aggfunc='sum')
groupby_data = groupby_data.sort_index()
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
groupby_data.head(10)
```

# Soal 6
---
- Dari soal nomor 5, kita ingin mencari top customer di jakarta.
- Cari data customer di area Jakarta yang total `box_qty`-nya di atas 200
- Simpan dalam variable `jakarta_data`
- *Hint: Gabungkan loc & conditional*

```python
# Tulis code disini
jakarta_data = groupby_data.loc['Jakarta']
jakarta_data = jakarta_data[jakarta_data['box_qty'] > 200]
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
jakarta_data
```

# Soal 7
---
- Buat kolom `price box_qty` dan `price total_weight` dari `data`
- Harga per `box_qty` adalah 10.000 dan harga per `weight` adalah 1.000.000
- Simpan dalam variable `data`
- *Hint*:
  - *Apabila ingin membuat kolom baru bernama `new` dari variable `data`, dapat gunakan sintaks berikut: `data['new'] = <isi kolom>`*
  - *Anda dapat melakukan operasi matematika dengan suatu kolom. Misal `data['berat'] * 3`*

```python
# Tulis code disini
data['price box_qty'] = data['box_qty'] * 10000
data['price total_weight'] = data['total_weight'] * 1000000
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
data.head()
```

# Soal 8
---
- Lanjutan dari soal 7, cari price setelah pajak apabila semua transaksi dikenakan pajak 10%.
- Buat kolom baru bernama `price_after_tax` untuk menyimpan harga setelah pajak
- Simpan dalam variable `data`
- *Hint*:
  - *Apabila ingin membuat kolom baru bernama `new` dari variable `data`, dapat gunakan sintaks berikut: `data['new'] = <isi kolom>`*
  - *Anda dapat melakukan operasi matematika dengan suatu kolom. Misal `data['berat'] * 3`*

```python
# Tulis code disini
data['total_price'] = data['price box_qty'] + data['price total_weight']
data['price_after_tax'] = data['total_price'] * 1.1 # pajak 10%
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
data.head()
```

# Soal 9
---
- Lanjutan dari soal 8, cari area mana yang jumlah price_after_tax-nya paling besar.
- Simpan dalam variable `top_area`

```python
# Tulis code disini
top_area = data.pivot_table(values='price_after_tax',
                            index='area',
                            aggfunc='sum')

# Sorting the result to get the area with the highest total
top_area = top_area.sort_values('price_after_tax', ascending=False).head(1)
```

```python
# JALANKAN CODE BERIKUT UNTUK VALIDASI
# JANGAN DIUBAH CODE-NYA
top_area
```
